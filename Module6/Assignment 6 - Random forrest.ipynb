```python
import pandas as pd
import numpy as np
import time
```

```python
# TODO: Load up the dataset into dataframe 'X'
df = pd.read_csv('C:/Users/acc_a/Desktop/DAT210x-master/Module6/Datasets/dataset-har-PUC-Rio-ugulino.csv', sep=';', decimal=",")
df.dtypes
df.shape
```

```python
df['z4'].unique()
```

```python
df = df.replace('-14420-11-2011 04:50:23.713', np.NaN)
df = df.dropna(axis=0)
df.shape
```

```python
df.head()
```

```python
df = df.replace(',', '.')
df['z4'] = df.z4.astype(float)
df.dtypes
```

```python
df['gender'] = pd.get_dummies(df['gender'], drop_first=True)
df.head()
```

```python
df['class'].unique()
```

```python
y = df[['class']]
y = pd.get_dummies(y, drop_first=False)
y.head()
```

```python
X = df.drop(['user', 'class'], axis=1)
X.head()
```

```python
X.describe()
```

```python
# TODO: Create an RForest classifier 'model' and set n_estimators=30,
# the max_depth to 10, and oob_score=True, and random_state=0
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=30, oob_score=True, max_depth=10, random_state=0)
```

```python
# TODO: Split your data into test / train sets
# Your test size can be 30% with random_state 7
# Use variable names: X_train, X_test, y_train, y_test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=7)
```

```python
print("Fitting...")
s = time.time()
#
# TODO: train your model on your training set
model.fit(X_train, y_train)
print("Fitting completed in: ", time.time() - s)
```

```python
# INFO: Display the OOB Score of your data
score = model.oob_score_
print("OOB Score: ", round(score*100, 3))
```

```python
print("Scoring...")
s = time.time()
#
# TODO: score your model on your test set
score= model.score(X_test, y_test)
print("Score: ", round(score*100, 3))
print ("Scoring completed in: ", time.time() - s)
```
